## Probability that Each Team Makes Each Round

```python
# Read in data (probability that each team wins in a game against another team)
df = pd.read_csv('March Madness.csv')
```

```python
# Find probability of making the Round of 32
r64_game_prob = {}
opps = {0: 1, 1: 0}
for n in range(32):
    prob_dict = {}
    teams = [df.loc[2*n + i, :][0] for i in range(2)]
    for i in range(2):
        r64_game_prob[teams[i]] = df.loc[2*n + i, :][2*n + opps[i] + 1]
print(r64_game_prob)
```

```python
# Find probability of making the Sweet Sixteen
r32_game_prob = {}
opps = {0: [2, 3], 1: [2, 3], 2: [0, 1], 3: [0, 1]}
for n in range(16):
    teams = [(df.loc[4*n + i, :][0], 4*n + i) for i in range(4)]
    for i in range(4):
        opp_name = [teams[idx][0] for idx in opps[i]]
        opp_idx = [teams[idx][1] for idx in opps[i]]
        r64_win_prob = r64_game_prob[teams[i][0]]
        opp_win_prob = [r64_game_prob[n] for n in opp_name]
        r32_game_prob[teams[i][0]] =  r64_win_prob*sum([df.loc[4*n + i, :][opp_idx[k] + 1]*opp_win_prob[k] for k in range(2)])
print(r32_game_prob)
```

```python
# Find probability of making the Elite Eight
r16_game_prob = {}
opps = {}
for i in range(4):
    opps[i] = [4, 5, 6, 7]
for i in range(4, 8):
    opps[i] = [0, 1, 2, 3]
for n in range(8):
    teams = [(df.loc[8*n + i, :][0], 8*n + i) for i in range(8)]
    for i in range(8):
        opp_name = [teams[idx][0] for idx in opps[i]]
        opp_idx = [teams[idx][1] for idx in opps[i]]  
        r32_win_prob = r32_game_prob[teams[i][0]]
        opp_win_prob = [r32_game_prob[n] for n in opp_name]
        r16_game_prob[teams[i][0]] =  r32_win_prob*sum([df.loc[8*n + i, :][opp_idx[k] + 1]*opp_win_prob[k] for k in range(4)])
print(r16_game_prob)
total = 0
for n in r16_game_prob:
    total += r16_game_prob[n]
print(total)
```

```python
# Find probability of making the Final Four
r8_game_prob = {}
opps = {}
for i in range(8):
    opps[i] = [8, 9, 10, 11, 12, 13, 14, 15]
for i in range(8, 16):
    opps[i] = [0, 1, 2, 3, 4, 5, 6, 7]

for n in range(4):
    teams = [(df.loc[16*n + i, :][0], 16*n + i) for i in range(16)]
    for i in range(16):
        opp_name = [teams[idx][0] for idx in opps[i]]
        opp_idx = [teams[idx][1] for idx in opps[i]]  
        r16_win_prob = r16_game_prob[teams[i][0]]
        opp_win_prob = [r16_game_prob[n] for n in opp_name]
        r8_game_prob[teams[i][0]] =  r16_win_prob*sum([df.loc[16*n + i, :][opp_idx[k] + 1]*opp_win_prob[k] for k in range(8)])
print(r8_game_prob)
total = 0
for n in r8_game_prob:
    total += r8_game_prob[n]
print(total)
```

```python
# Find probability of making the final
r4_game_prob = {}
opps = {}
for i in range(16):
    opps[i] = [i + 16 for i in range(16)]
for i in range(16, 32):
    opps[i] = [i for i in range(16)]

for n in range(2):
    teams = [(df.loc[32*n + i, :][0], 32*n + i) for i in range(32)]
    for i in range(32):
        opp_name = [teams[idx][0] for idx in opps[i]]
        opp_idx = [teams[idx][1] for idx in opps[i]]  
        r8_win_prob = r8_game_prob[teams[i][0]]
        opp_win_prob = [r8_game_prob[n] for n in opp_name]
        r4_game_prob[teams[i][0]] =  r8_win_prob*sum([df.loc[32*n + i, :][opp_idx[k] + 1]*opp_win_prob[k] for k in range(16)])
print(r4_game_prob)
total = 0
for n in r4_game_prob:
    total += r4_game_prob[n]
print(total)
```

```python
# Find probability of winning the whole tournament
r2_game_prob = {}
opps = {}
for i in range(32):
    opps[i] = [i + 32 for i in range(32)]
for i in range(32, 64):
    opps[i] = [i for i in range(32)]

teams = [(df.loc[i, :][0], i) for i in range(64)]
for i in range(64):
    opp_name = [teams[idx][0] for idx in opps[i]]
    opp_idx = [teams[idx][1] for idx in opps[i]]  
    r4_win_prob = r4_game_prob[teams[i][0]]
    opp_win_prob = [r4_game_prob[n] for n in opp_name]
    r2_game_prob[teams[i][0]] =  r4_win_prob*sum([df.loc[i, :][opp_idx[k] + 1]*opp_win_prob[k] for k in range(32)])
print(r2_game_prob)
total = 0
for n in r2_game_prob:
    total += r2_game_prob[n]
print(total)
```

```python
# Formulate final probabilities
probs = {}
probs[1] = r64_game_prob
probs[2] = r32_game_prob
probs[3] = r16_game_prob
probs[4] = r8_game_prob
probs[5] = r4_game_prob
probs[6] = r2_game_prob
probs
```

```python
# Convert probabilities into a dataframe to use
p = pd.DataFrame(probs)
```

## Integer Programming Formulation

Let's define the variable that holds the boolean value of if a team i won match j as $x_{i, j}$, with n and m being the total number of teams and matches, respectively. We subtract one from the lower and upper bound of i and j to match the indexing used in Python. Then we have the following constraints:

$$
\begin{aligned}
max \sum_{i=0}^{n-1}\sum_{j=0}^{m-1} 2^{j} \cdot x_{i,j} \cdot p_{i,j}\\
\end{aligned}
$$
$$
\begin{aligned}
s.t. \ x_{i,j} \le x_{i,j+1} \ \forall i \in \{0,n-1\} \ \forall j \in \{0,m-2\}\\
x_{2i,0} + x_{2i+1,0} = 1 \ \forall i \in \{0,\frac{n}{2}-1\}\\
x_{4i,1} + x_{4i+1,1} + ... + x_{4i+3,1} = 1 \ \forall i \in \{0,\frac{n}{4}-1\}\\
x_{8i,2} + x_{8i+1,2} + ... + x_{8i+7,2} = 1 \ \forall i \in \{0,\frac{n}{8}-1\}\\
x_{16i,3} + x_{16i+1,3} + ... + x_{16i+15,3} = 1 \ \forall i \in \{0,\frac{n}{16}-1\}\\
x_{32i,4} + x_{32i+1,4} + ... + x_{32i+31,4} = 1 \ \forall i \in \{0,\frac{n}{32}-1\}\\
\sum_{i=1}^n x_{i,5} = 1\\
x_{i,j} \in \{0,1\} \ \forall i \in \{1,n-1\} \ \forall j \in \{1,m-1\}
\end{aligned}
$$


## Integer Programming Implementation

```python
# Define model and objective

m = Model()

x = m.addVars(64, 6, lb=0, name="x", vtype=GRB.BINARY)

m.setObjective(quicksum(p.iloc[i, j] * x[i, j] * 2**j for i in range(64) for j in range(6)), GRB.MAXIMIZE)
```

```python
# Define constraints

m.addConstrs(x[i, j] >= x[i, j + 1] for i in range(64) for j in range(5))
m.addConstrs(quicksum(x[2 * i + k, 0] for k in range(2)) == 1 for i in range(32))
m.addConstrs(quicksum(x[4 * i + k, 1] for k in range(4)) == 1 for i in range(16))
m.addConstrs(quicksum(x[8 * i + k, 2] for k in range(8)) == 1 for i in range(8))
m.addConstrs(quicksum(x[16 * i + k, 3] for k in range(16)) == 1 for i in range(4))
m.addConstrs(quicksum(x[32 * i + k, 4] for k in range(32)) == 1 for i in range(2))
m.addConstr(quicksum(x[i, 5] for i in range(64)) == 1)
```

```python
# Run model and add objective value to expected points list

m.optimize()
exp_pts.append(m.objVal)

print("#############################################################################")
print(f"\nThe optimal objective value is {m.objVal}.")
```

```python
# Format optimal solution

final_matrix = []
for i in range(len(m.getVars())):
    if i % 6 == 0:
        inner = []
    inner.append(m.getVars()[i].x)
    if i % 6 == 0:
        final_matrix.append(inner)

final_matrix = np.array(final_matrix).reshape(64, 6).astype(int)
matrix_list.append(final_matrix)
final_matrix
```

```python
bracket = pd.DataFrame({"Team": p.index, "Number of Wins": final_matrix.sum(axis=1).astype(int)})
bracket_list.append(bracket)
pprint(bracket)
```

## Formulation of Finding 2nd best model

To find the second-best model, we introduce the constraint that the optimal solution cannot be identical to the optimal solution of the best model. Let's define the matrix that holds the best model as A and the variable that holds the boolean value of if a team i won match j as $x_{i, j}$, with n and m being the total number of teams and matches, respectively. Then we add the following constraint to our model:

$\sum_{j=0}^{m-1}\sum_{i=0}^{n-1} x_{i, j} \cdot A_{i, j} - \sum_{j=0}^{m-1}\sum_{i=0}^{n-1} x_{i, j} \cdot (1 - A_{i, j}) \le \sum_{j=0}^{m-1}\sum_{i=0}^{n-1} (A_{i, j}) - 1$

## Implementation of Finding 2nd best model

```python
# Define model and objective

m = Model()

x = m.addVars(64, 6, lb=0, name="x", vtype=GRB.BINARY)

m.setObjective(quicksum(p.iloc[i, j] * x[i, j] * 2**j for i in range(64) for j in range(6)), GRB.MAXIMIZE)
```

```python
# Define constraints with additional constrant that prevents the best solution from being outputted

m.addConstrs(x[i, j] >= x[i, j + 1] for i in range(64) for j in range(5))
m.addConstrs(quicksum(x[2 * i + k, 0] for k in range(2)) == 1 for i in range(32))
m.addConstrs(quicksum(x[4 * i + k, 1] for k in range(4)) == 1 for i in range(16))
m.addConstrs(quicksum(x[8 * i + k, 2] for k in range(8)) == 1 for i in range(8))
m.addConstrs(quicksum(x[16 * i + k, 3] for k in range(16)) == 1 for i in range(4))
m.addConstrs(quicksum(x[32 * i + k, 4] for k in range(32)) == 1 for i in range(2))
m.addConstr(quicksum(x[i, 5] for i in range(64)) == 1)

# New constraint
m.addConstr(quicksum(x[i, j] * final_matrix[i, j] for i in range(64) for j in range(6)) - quicksum(x[i, j] * 
(1-final_matrix[i, j]) for i in range(64) for j in range(6)) <= np.sum(final_matrix) - 1)
```

```python
# Run model

m.optimize()
exp_pts.append(m.objVal)

print("#############################################################################")
print(f"\nThe optimal objective value is {m.objVal}.")
```

# Alternative Model to Improve Actual Performance





The first method to improve upon the IP in part B is to implement a constraint to have one of each seed win more than the average number of games that the seed wins in March Madness historically. This essentially adds in a constraint to take into account historical trends of March Madness. The actual score increases from 78 points to 126 points even though the new model is sub-optimal for the inital IP. This is because there will most definitely be deviations from the most likely scenario (the scenario that is accounted for in part b). Essentially, this method uses historical trends to guess what these deviations will be.

```python
# Define model and objective

m = Model()

x = m.addVars(64, 6, lb=0, name="x", vtype=GRB.BINARY)
avg_wins_by_seed = {0: 3, 1: 2, 2: 1, 3: 1, 4: 1, 5:1, 6:0, 7:0, 8:0, 9:0, 10:0, 11:0, 12:0, 13:0, 14:0, 15:0}

m.setObjective(quicksum(p.iloc[i, j] * x[i, j] * 2**(j) for i in range(64) for j in range(6)), GRB.MAXIMIZE)

# Define constraints

m.addConstrs(x[i, j] >= x[i, j + 1] for i in range(64) for j in range(5))
m.addConstrs(quicksum(x[2 * i + k, 0] for k in range(2)) == 1 for i in range(32))
m.addConstrs(quicksum(x[4 * i + k, 1] for k in range(4)) == 1 for i in range(16))
m.addConstrs(quicksum(x[8 * i + k, 2] for k in range(8)) == 1 for i in range(8))
m.addConstrs(quicksum(x[16 * i + k, 3] for k in range(16)) == 1 for i in range(4))
m.addConstrs(quicksum(x[32 * i + k, 4] for k in range(32)) == 1 for i in range(2))

# At least one team of each seed has to win more than the average number of wins for each seed
m.addConstrs(x[s, avg_wins_by_seed[s]] + x[s + 16, avg_wins_by_seed[s]] + x[s + 32, avg_wins_by_seed[s]] + 
             x[s + 48, avg_wins_by_seed[s]] >= 1 for s in range(16))

# Run model and add objective value to expected points list

m.optimize()

# Format optimal solution

final_matrix = []
for i in range(len(m.getVars())):
    if i % 6 == 0:
        inner = []
    inner.append(m.getVars()[i].x)
    if i % 6 == 0:
        final_matrix.append(inner)

final_matrix = np.array(final_matrix).reshape(64, 6).astype(int)
matrix_list.append(final_matrix)

bracket = pd.DataFrame({"Team": p.index, "Number of Wins": final_matrix.sum(axis=1).astype(int)})

print("#############################################################################")
print(f"\nThe optimal objective value is {m.objVal}.")
print(f"\nThe number of points scored for this bracket is {calc_score(bracket)}.")
```
